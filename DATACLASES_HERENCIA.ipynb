```python
from dataclasses import dataclass

@dataclass
class libro:
    _titulo: str
    _autor: str
    _isbn: str
    _precio: float

    # ---------------------------------------------
    #GETTERS & SETTERS - API publica limpia y estandar
    # ----------------

    @property #SON GETTERS
    def titulo(self) -> str:
        """Obtiene el título del libro."""
        return self._titulo

    @property
    def autor(self) -> str:
        """Obtiene el autor del libro."""
        return self._autor

    @property
    def isbn(self) -> str:
        """Obtiene el ISBN del libro."""
        return self._isbn

    @property
    def precio(self) -> float:
        """Obtiene el precio del libro."""
        return self._precio

    @titulo.setter #SON SETTERS
    def titulo(self, valor: str) -> None:
        """Establece el título del libro con validación."""
        if isinstance(valor, str) and valor.strip():
            self._titulo = valor
        else:
            raise ValueError("El título debe ser un texto válido no vacío.")


    @autor.setter
    def autor(self, valor: str) -> None:
        """Establece el autor del libro con validación."""
        if isinstance(valor, str) and valor.strip():
            self._autor = valor
        else:
            raise ValueError("El autor debe ser un texto válido no vacío.")

    @isbn.setter
    def isbn(self, valor: str) -> None:
        """Establece el ISBN del libro con validación."""
        if isinstance(valor, str) and valor.strip():
            self._isbn = valor
        else:
            raise ValueError("El ISBN debe ser un texto válido no vacío.")


    @precio.setter
    def precio(self, valor: float) -> None:
        """Establece el precio del libro con validación."""
        if isinstance(valor, (int, float)) and valor >= 0:
            self._precio = valor
        else:
            raise ValueError("El precio debe ser un número positivo.")


    def __repr__(self) -> str:
        """Representación clara del objeto Libro."""
        return (
            f"Libro(título='{self._titulo}', autor='{self._autor}',"
            f" ISBN='{self._isbn}', precio={self._precio})"
            )

def main() -> None:
    my_libro = libro("El Quijote", "Miguel de Cervantes", "978-3-16-148410-0", 19.99)

    print("\nInformación del libro:")
    print(my_libro)

    # Actualizamos el precio utilizando el setter
    my_libro.precio = 24.99
    my_libro.titulo = "Don Quijote de la Mancha"  # Actualizamos el título utilizando el setter

    print("\nInformación actualizada del libro:")
    print(my_libro)

if __name__ == "__main__":
    main()
```
