```python
class Prendas: #clase
    def __init__(self, nombre, precio, stock, color, ubicacion): #constructor
        self.__nombre = nombre #atributos
        self.__precio = precio
        self.__stock = stock
        self.__color = color
        self.__ubicacion = ubicacion
    def get_precio(self):  #metodo, devolver el  valor contenido en el atributo precio
     return self.__precio

    def get_stock(self):
     return self.__stock

    def get_ubicacion(self):
     return self.__ubicacion


    def set_precio(self, nuevo_precio): #metodo set para modificar  un  valor o actualizarlo
        if isinstance(nuevo_precio, (int, float)) and nuevo_precio >= 0: #validar que el nuevo precio sea un numero positivo
         print("El precio actual es:  ", self.__precio)
         self.__precio = nuevo_precio
         print("El nuevo precio es de: ",nuevo_precio)
        else:
         print("El precio debe ser un número positivo.")

    def set_stock(self, nuevo_stock):
      if isinstance(nuevo_stock, (int, float)) and nuevo_stock >= 0:
        print("El stock actual es: ", self.__stock)
        self.__stock = nuevo_stock
        print("El nuevo stock es de: " , nuevo_stock)
      else:
          print("El stock debe ser positivo. ")

    def set_ubicacion(self, nueva_ubicacion):
      if isinstance(nueva_ubicacion, (int , float)) and nueva_ubicacion >= 0:
        print("La ubicacion actual es: ", self.__ubicacion)
        self.__ubicacion = nueva_ubicacion
        print("La nueva ubicacion es: ", nueva_ubicacion)
      else:
          print("La ubicacion debe ser positiva. ")





    def mostrar_informacion(self): #metodo para mostrar la informacion las prendas
     print(f"nombre: {self.__nombre}")
     print(f"Precio: {self.__precio:,.2f}")
     print(f"stock: {self.__stock}")
     print(f"color: {self.__color}")
     print(f"ubicacion: {self.__ubicacion}")

def main():
    print("\nSISTEMA DE PRENDAS\n")
    prenda1 = Prendas("CAMISAS", 32.500 , 15 , "ROJAS" ,356899)
    prenda2 = Prendas("PANTALONES", 100.000, 12 , "AZULES" , 246638)
    prenda3 = Prendas("CHAQUETAS", 25.999 , 18 ,"NEGRAS" , 51242638)
    prenda4 = Prendas("MEDIAS", 54.999 , 35 ,"COLORES VARIADOS", 324617)

    print("\nLOS DATOS DE LAS PRENDAS SON")
    print("\nLOS DATOS DE LA PRIMERA PRENDA SON ")
    prenda1.mostrar_informacion()

    print("\nLOS DATOS DE LA SEGUNDA PRENDA SON")
    prenda2.mostrar_informacion()

    print("\nLOS DATOS DE LA TERCERA PRENDA SON ")
    prenda3.mostrar_informacion()

    print("\nLOS DATOS DE LA CUARTA PRENDA SON")
    prenda4.mostrar_informacion()


#Vamos a modificar los datos de la prendas
    print("\nMODIFICACION DE LA PRIMERA PRENDA")
    prenda1.set_precio(-45.000)
    prenda1.set_precio(50.000)
    prenda1.set_stock(-14)
    prenda1.set_stock(20)
    prenda1.set_ubicacion(-17)
    prenda1.set_ubicacion(35)


    print("\nMODIFICACION DE LA SEGUNDA PRENDA")
    prenda2.set_precio(-60.000)
    prenda2.set_precio(36.000)
    prenda2.set_stock(-19)
    prenda2.set_stock(40)
    prenda2.set_ubicacion(-12)
    prenda2.set_ubicacion(52)



    print("\nMODIFICACION DE LA TERCERA PRENDA")
    prenda3.set_precio(-42.999)
    prenda3.set_precio(32.000)
    prenda3.set_stock(-22)
    prenda3.set_stock(16)
    prenda3.set_ubicacion(-33)
    prenda3.set_ubicacion(25)


    print("\nMODIFICACION DE LA CUARTA PRENDA")
    prenda4.set_precio(-43.990)
    prenda4.set_precio(62.999)
    prenda4.set_stock(-25)
    prenda4.set_stock(15)
    prenda4.set_ubicacion(-10)
    prenda4.set_ubicacion(37)



 #Vamos a mostrar el nuevo precio actual de las prendas
    print("\nNUEVO PRECIO ACTUAL DE LA PRIMERA PRENDA.")
    print("EL PRECIO NUEVO DE LA PRIMERA PRENDA  ES: ", prenda1.get_precio())
    print("EL STOCK NUEVO DE LA PRIMERA PRENDA ES: ", prenda1.get_stock())
    print("LA UBICACION NUEVA DE LA PRIMERA PRENDA  ES: ", prenda1.get_ubicacion())
    prenda1.mostrar_informacion()

    print("\nNUEVO PRECIO ACTUAL DE LA SEGUNDA PRENDA.")
    print("EL PRECIO NUEVO DE LA SEGUNDA PRENDA  ES: ", prenda2.get_precio())
    print("EL STOCK NUEVO DE LA SEGUNDA PRENDA  ES: ", prenda2.get_stock())
    print("LA UBICACION NUEVA DE LA SEGUNDA PRENDA  ES: ", prenda2.get_ubicacion())
    prenda2.mostrar_informacion()

    print("\nNUEVO PRECIO ACTUAL DE LA TERCERA PRENDA.")
    print("EL PRECIO NUEVO DE LA TERCERA PRENDA  ES: ", prenda3.get_precio())
    print("EL STOCK NUEVO DE LA TERCERA PRENDA  ES: ", prenda3.get_stock())
    print("LA UBICACION NUEVA DE LA TERCERA PRENDA  ES: ", prenda3.get_ubicacion())
    prenda3.mostrar_informacion()

    print("\nNUEVO PRECIO ACTUAL DE LA CUARTA PRENDA")
    print("EL PRECIO NUEVO DE LA CUARTA PRENDA  ES: ", prenda4.get_precio())
    print("EL STOCK NUEVO DE LA CUARTA PRENDA  ES: ", prenda4.get_stock())
    print("LA UBICACION NUEVA DE LA CUARTA PRENDA  ES: ", prenda4.get_ubicacion())
    prenda4.mostrar_informacion()



    print("FIN DE LOS DATOS DE LAS PRENDAS.")

if __name__ == "__main__":
    main()
```
