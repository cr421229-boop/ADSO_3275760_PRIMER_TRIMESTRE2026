```python
#El ambiente 218 desea guardar la lista de nombres de los elementos
#El sistema debe hacer:
#1. Agregar nuevos muebles append()
#2. Recorrer la lista y mostrar los muebles for
#3. Modificar un nombre en caso de error
#4. Borrar un mueble

#==============================
#SOLUCION
#===============================


#1.Agregar nuevos elementos append()

def agregar_elemento(lista_elementos, nombre):
    if isinstance(nombre, str) and 2 <= len(nombre) <= 50:
        lista_elementos.append(nombre)
        print(f"\nELEMENTO AGREGADO: {nombre}")

    else:
        print("\nNOMBRE DE ELEMENTO INVALIDO DEBE TENER ENTRE 2 A 50 CARACTERES")

#2. Recorrer la lista y mostrar los elementos for
def mostrar_elemento(lista_elementos):
    for elemento in lista_elementos:
        print(f"-> {elemento}")

#3.Modificar un nombre en caso de error
def  modificar_elemento(lista_elementos, indice , nuevo_nombre):
    if isinstance(nuevo_nombre, str) and 2 <= len(nuevo_nombre) <= 50:
        print("\nNOMBRE DE  ELEMENTO INVALIDO DEBE TENER ENTRE 2 A 50 CARACTERES")
        return #Termina la funcion
    if 0 <= indice < len(lista_elementos):
        nombre_original = lista_elementos[indice]
        lista_elementos[indice] = nuevo_nombre.title()
        print(f"\nEL ELEMENTO {nombre_original} FUE REEMPLAZO POR: {nuevo_nombre.title()}")
    else:
        print("\nNO SE ENCONTRO EL DATO, NUMERO DE LA LISTA ERRONEO")

#4Borrar  un elemento
def borrar_elemento(lista_elementos, indice):
    if 0 <= indice < len(lista_elementos):
        borrar = lista_elementos.pop(indice)
        print(f"\nEL ELEMENTO {borrar}, SE BORRO DE LA LISTA DE CLIENTES")
    else:
        print("\nNO SE ENCONTRO EL ELEMENTO, ELEMENTO NO BORRADO DE LA LISTA.")

def main():
    #Simulacion empresarial
   elementos =  ["COMPUTADORES", "SILLAS", "MESAS", "TABLERO","VIDEOBEAM","LAMPARAS","CORTINAS","VITRINA"]
   print("\nELEMENTOS ACTUALES")
   mostrar_elemento(elementos)
   #Agregar nuevos elementos  append()
   agregar_elemento(elementos, "PUERTA")
   mostrar_elemento(elementos)
   #Modificar un nombre en caso de error
   modificar_elemento(elementos,1,"COMPUTADOR")
   mostrar_elemento(elementos)
   #Borrar un elemento
   borrar_elemento(elementos, 3)
if __name__ == "__main__":
    main()
```
