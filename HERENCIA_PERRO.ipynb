```python
class Vehiculo:
    def mover(self):
        print("\nTODOS LOS VEHICULOS SE MUEVEN A SU MANERA")

class Moto(Vehiculo):
    def mover(self):
        print("\nLA MOTO ESTA CONDUCIENDO")
class Carro(Vehiculo):
    def mover(self):
        print("\nEL CARRO CONDUCE EN LA CARRETERA")
class Avion(Vehiculo):
    def mover(self):
        print("\nEL AVION ESTA VOLANDO EN EL AIRE")
class Submarino(Vehiculo):
    def mover(self):
        print("\nEL SUBMARINO ESTA NAVEGANDO EN EL AGUA")
class Barco(Vehiculo):
    def mover(self):
        print("\nEL BARCO ESTA NAVEGANDO EN EL AGUA")
def main():
    vehiculo1 = Vehiculo()
    vehiculo1.mover()
    moto1 = Moto()
    moto1.mover()
    carro1 = Carro()
    carro1.mover()
    avion1 = Avion()
    avion1.mover()
    barco1 = Barco()
    barco1.mover()
if __name__ == "__main__":
    main()
```
