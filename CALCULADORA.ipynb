```python
class operacion:
    def __init__(self,num1,num2):
        self.num1 = num1
        self.num2 = num2
    def calcular (self):
        print("Operaciones genericos")

class suma(operacion):
    def calcular(self):

        resultado= self.num1 + self.num2
        print(f"LA SUMA ES: {resultado}")
        print("\n-------FIN DE SUMA-------")

class resta(operacion):
    def calcular(self):
        resultado=int(self.num1 - self.num2)
        print(f"LA RESTA ES: {resultado}")
        print("\n-------FIN DE RESTA-------")
class multiplicacion(operacion):
    def calcular(self):
        resultado=int(self.num1 * self.num2)
        print(f"LA MULTIPLICACION ES: {resultado}")
        print("\n-------FIN DE MULTIPLICACION-------")
class division(operacion):
    def calcular(self):
        if self.num2 == 0:
            print("Error: No se puede dividir por cero")
        else:
            resultado=int (self.num1/self.num2)
            print(f"LA DIVISION ES: {resultado}")
            print("\n-------FIN DE DIVISION-------")
class potencia(operacion):
    def calcular(self):
        if self.num2 > 3:
            print("Error: No se puede potenciar a mas de tres")
        else:
            resultado=float (self.num1**self.num2)
            print(f"LA POTENCIA ES: {resultado}")
            print("\n-------FIN DE POTENCIA-------")

def main():
    numero1= float(input("\nIngrese el primer valor: "))
    numero2= float(input("Ingrese el segundo valor: "))
    objeto1= suma(numero1,numero2)
    objeto1.calcular()
    numero1= float(input("\nIngrese el primer valor: "))
    numero2= float(input("Ingrese el segundo valor: "))
    objeto1= resta(numero1,numero2)
    objeto1.calcular()
    numero1= float(input("\nIngrese el primer valor: "))
    numero2= float(input("Ingrese el segundo valor: "))
    objeto1= multiplicacion(numero1,numero2)
    objeto1.calcular()
    numero1= float(input("\nIngrese el primer valor: "))
    numero2= float(input("Ingrese el segundo valor: "))
    objeto1= division(numero1,numero2)
    objeto1.calcular()
    numero1= float(input("\nIngrese el primer valor: "))
    numero2= float(input("Ingrese el segundo valor: "))
    objeto1= potencia(numero1,numero2)
    objeto1.calcular()
if __name__ == "__main__":
    main()
```
