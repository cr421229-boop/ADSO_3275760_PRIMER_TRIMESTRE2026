```python
class Vehiculo:
    def mover(self):
        print("\nEl vehiculo se esta moviendo")

class Carro(Vehiculo):
    def mover(self): #AQUI HACEMMOS LA SOBRESCRITURA DEL METODO MOVER ()
        print("\nEl carro se esta moviendo")


class Avion(Vehiculo):
    def mover(self): #AQUI HACEMOS LA SOBRESCRITURA DEL METODO MOVER ()
        print("\nEl avion esta volando")

class Barco(Vehiculo):
    def mover(self): #AQUI HACEMOS LA SOBRESCRITURA DEL METODO MOVER ()
        print("\nEl barco se esta hundiendo")


class Submarino(Vehiculo):
    def mover(self): #AQUI HACEMOS LA SOBRESCRITURA DEL METODO MOVER ()
        print("\nEl submarino se esta sumergiendo")

def main():
    vehiculo1 = Vehiculo() #Creo un objeto de la clase padre
    vehiculo1.mover() #Llamo al metodo mover del objeto de la clase


    #Aqui nos muestra sel sobrescritura del metodo mover() de la clase carro
    Carro1 = Carro() #Creo un objeto de la clase hija
    Carro1.mover() #Llamo al metodo mover del objeto de la clase hija


    Avion1 = Avion() #Creo un objeto de la clase hija
    Avion1.mover() #Llamo al metodo mover del objeto de la clase


    Barco1 = Barco() #Creo un objeto de la clase hija
    Barco1.mover() #Llamo al metodo mover del objeto de la clase

    Submarino1 = Submarino() #Creo un objeto de la clase hija
    Submarino1.mover() #Llamo al metodo mover del objeto de la clase



if __name__ == "__main__":
    main()
```
